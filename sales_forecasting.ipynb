```python
data=pd.read_csv('C:/Users/malte/OneDrive/Desktop/Sujay/advertising.csv')
```

```python
data.head()
```

```python
import statsmodels.api as sm
```

```python
X = data[['TV', 'Radio', 'Newspaper']]
```

```python
y = data['Sales']
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())
```

90.3% of the variation in Sales is explained by TV, Radio, and Newspaper ads. Very strong fit.

Adjusted for number of predictors. Still very high â†’ model is not overfitting.

The overall model is statistically significant.
